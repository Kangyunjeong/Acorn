```python
from selenium import webdriver
```

```python
# 웹드라이버 로딩
driver = webdriver.Chrome()
```

```python
# 페이지 요청
driver.get('http://www.11st.co.kr/html/bestSellerMain.html')
```

```python
# 필요한 selector 모두 가져오기
li_list = driver.find_elements_by_css_selector('#bestPrdList > div.viewtype.catal_ty > ul > li')
li_list
```

```python
# seletor 잘 들어왔는지 확인
li_list[0].text
```

```python
# 데이터 넣을 빈 딕셔너리 생성
data_dic = {
    'name' : [],
    'price' : [],
    'point' : [],
    'shipping' : [],
    'mall' : []
}

# li_list의 각 tag를 이름, 가격, 포인트, 배송비, 쇼핑몰 별로 정리
for li_tag in li_list :
    name_tag = li_tag.find_element_by_css_selector('#bestPrdList > div.viewtype.catal_ty > ul > li > div > a > div.pname > p')
    price_tag = li_tag.find_element_by_css_selector('#bestPrdList > div.viewtype.catal_ty > ul > li > div > a > div.pname > div.price_info.cfix > span > strong')
    
    # 태그에 값이 없는 경우 예외처리
    try:
        point_tag = li_tag.find_element_by_css_selector('#bestPrdList > div.viewtype.catal_ty > ul > li > div > a > div.pname > div.prd_flag > em > span.point2.lsn')
    except:
        pass
    
    # 태그에 값이 없는 경우 예외처리
    try:
        shipping_tag = li_tag.find_element_by_css_selector('#bestPrdList > div.viewtype.catal_ty > ul > li > div > a > div.pname > div.prd_flag > div > em')
    except:
        pass
        
    mall_tag = li_tag.find_element_by_css_selector('#bestPrdList > div.viewtype.catal_ty > ul > li > div > div.store > a')
    
    print(name_tag.text)
    print(price_tag.text)
    print(point_tag.text)
    print(shipping_tag.text)
    print(mall_tag.text)
    
    # 딕셔너리에 텍스트값 저장
    data_dic['name'].append(name_tag.text)
    data_dic['price'].append(price_tag.text)
    data_dic['point'].append(point_tag.text)
    data_dic['shipping'].append(shipping_tag.text)
    data_dic['mall'].append(mall_tag.text)
```

```python
import pandas

df1 = pandas.DataFrame.from_dict(data_dic)
df1.to_csv('11번가_크롤링.csv', encoding='utf-8-sig', index=False)
print('저장완료')
```
